```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
# TODO

df = pd.read_csv("videogamesales.csv")
df.head()
```

```python
display(df.shape)
display(df.info())
```

```python
df['Genre'].unique()
```

```python
#Actual Assignment

df['global_sales'] = df['EU_Sales']+df['JP_Sales']+df['NA_Sales']+df['Other_Sales']
df_genre = df.groupby('Genre')['global_sales'].sum().sort_values(ascending=False)
display(df_genre)

df_genre.plot(kind='bar', title='game Sales Across Genres', xlabel='Genre', ylabel='Sales')
```

```python
df['EU_JP_sales']=df['EU_Sales']+df['JP_Sales']
df_GTA = df[['Name', 'Platform', 'Year','EU_JP_sales']][df['Name'].str.contains('Grand Theft Auto')]
display(df_GTA)
```

```python
df_gta = df[['Name','global_sales']][df["Name"].str.contains('Grand Theft Auto')]
df_gta = df_gta.set_index('Name')['global_sales']

df_gta.plot(kind='pie', title='Grand Theft Auto', y='global_sales', x='Name', legend=True, figsize=(20,20))
```
